## <font color='yellow'>Regressão</font>

### <font color='yellow'>Métrica para Avaliação de Modelos de Regressão</font>

Avaliar o desempenho de um modelo de regressão é essencial para entender o quão bem ele está prevendo os valores de uma variável dependente (target). Avaliar o desempenho preditivo de um modelo de regressão de forma simples para um conjunto de objetos é calcular a soma das diferenças entre o valor predito e o valor correto. O desempenho do modelo é avaliado pela soma dessas diferenças. Para ilustrar esse procedimento, suponha que para os 5 objetos, **valor predito**, **valor verdadeiro**, e a respectiva **diferença**, sejam dados pela Tabela abaixo.

\

| Valor Predito ($f(x_i)$) | Valor verdadeiro ($y_i$) | Diferença entre os valores |
|---------------|------------|----------------------------|
| 0,9           | 0,6        | 0,3                        |
| 0,6           | 0,8        | -0,2                       |
| 0,4           | 0,1        | 0,3                        |
| 0,8           | 0,7        | 0,1                        |
| 0,5           | 0,9        | -0,4                       |

\

É possível observar que a soma das diferenças, no caso 0,1, não reflete a distância real entre os vaores preditos e seus respectivos valores verdadeiros. Isso porque, como temos valores positivos e negativos, valores de diferenças negativas reduzem valores de diferenças positivas.

Isso pode ser evitado elevando os valores das diferenças ao quadrado, o que é feito pela medida da **Soma dos Quadrados dos Erros** (**SSE, *Sum of Squared Error***), como mostra a seguinte Equação:

\

$$
\displaystyle
\text{SSE} = \sum_{i=1}^{n} (y_i - f(x_i))^2
$$

\

Nesse medida, quanto menor o valor, menor o erro na predição do valor verdadeiro, e, assim, melhor é o desempenho do modelo.


Um problema do **SSE** é que seu valor depende do número de objetos para os quais o desempenho está sendo avaliado. Isso pode ser evitado utilizando a média do quadrado das diferenças, o que é feito pela **Média dos Quadrados dos Erros** (**MSE, *Mean Squared Error***), como mostrado abaixo:

\

$$
\displaystyle
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - f(x_i))^2
$$

\


É importante observar que, para o caso do **SSE** e do **MSE**, ao elevar as diferenças ao quadrado, torna-se difícil a interpretação do erro. Existem, no entanto, duas formas simples de resolver esse problema:

1. Tirar a raiz quadrada do resultado, o que gera a métrica chamada de **Raiz do Erro Quadrático Médio** (**RMSE, *Root Mean Square Error***), conforme ilustrado abaixo:

\

$$
\displaystyle
\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - f(x_i))^2}
$$

\

2. Em vez de elevar cada diferença ao quadrado, usar o módulo de cada diferença, calculando, assim, a **Média dos Valores Absolutos** ou o **Erro Médio Absoluto** (**MAE, *Mean Absolute Error***). Além de mais simples, essa métrica evita um viés da alternativa anterior, que é, ao elevar as diferenças ao quadrado, fazer com que as maiores diferenças se sobressaiam com relação às outras. A Equação para calcular a **MAE**:

\

$$
\displaystyle
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - f(x_i)|
$$

\

Essas duas medidas, por terem a mesma unidade de medida que o valor a ser predito $y_i$, tornam-se mais fáceis de interpretar do que o SSE e o MSE.

Para exemplificar o uso das métricas de regressão, dois estudos de caso serão apresentados e em cada um deles avaliaremos o desempenho do modelo treinado com diferentes métricas.



#### Um Resumo

as métricas mais utilizadas na avaliação de modelos de regressão são:

1. **MAE (Mean Absolute Error)**
2. **MSE (Mean Squared Error)**
3. **RMSE (Root Mean Squared Error)**
4. **R² Score**

Cada métrica tem características específicas e fornece diferentes perspectivas sobre o desempenho do modelo.







---

#### 1. Mean Absolute Error (MAE)

-  Definição

O **Mean Absolute Error** é a média dos erros absolutos entre os valores preditos e os valores verdadeiros:

$$
\displaystyle
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - f(x_i)|
$$

\

- Características

1. Mede a magnitude média dos erros sem considerar a direção (positiva ou negativa).
2. Quanto menor o MAE, melhor o desempenho do modelo.
3. Sensível à escala dos dados.

\

- Vantagens
1.  Interpretação direta: expressa o erro na mesma unidade da variável dependente.
2. Menos sensível a outliers comparado ao MSE.

\

- Desvantagens
1. Pode não penalizar suficientemente erros grandes.

\

- Exemplo Prático

Se um modelo está prevendo o preço de casas e o MAE é 15.000, significa que, em média, o modelo erra o preço em 15.000 unidades monetárias.


#### 2. Mean Squared Error (MSE)

- Definição

O **Mean Squared Error** é a Média dos Quadrados dos Erros:

$$
\displaystyle
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - f(x_i))^2
$$

\

- Características

1. Penaliza mais fortemente erros grandes devido ao termo ao quadrado.
2. Quanto menor o MSE, melhor o desempenho do modelo.

\

- Vantagens
1. Destaca erros grandes, o que pode ser útil em aplicações onde esses erros são mais impactantes.

\

- Desvantagens
1. Menos intuitivo porque está em uma escala diferente da variável dependente (unidades ao quadrado).
2. Sensível a outliers.

\

- Exemplo Prático

Se o MSE é 225, significa que o erro quadrático médio é 225 unidades ao quadrado, exigindo interpretação cuidadosa.




#### 3. Root Mean Squared Error (RMSE)

- Definição

O **Root Mean Squared Error** é a raiz quadrada do MSE:

$$
\displaystyle
\text{RMSE} = \sqrt{ \frac{1}{n} \sum_{i=1}^{n} (y_i - f(x_i))^2 }
$$

\

- Características
1. Mede o erro em uma escala similar à variável dependente.
2. Penaliza erros grandes de forma similar ao MSE.

\

- Vantagens
1. Mais interpretável que o MSE, pois é expresso na mesma unidade da variável dependente.

\

- Desvantagens
1. Ainda sensível a outliers, como o MSE.

- Exemplo Prático

Se o RMSE é 15, isso significa que, em média, a magnitude do erro é 15 unidades na mesma escala da variável alvo.


#### 4. R² Score (Coeficiente de Determinação)

- Definição

O **R² Score** mede a proporção da variância explicada pelo modelo em relação à variância total dos dados:

$$
\displaystyle
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - f(x_i))^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
$$

\

Onde:
- $y_i$: Valor verdadeiro da variável dependente no exemplo $i$.

- $f(x_i)$: Valor predito pelo modelo para o exemplo $i$.

- $\bar{y}$: Média de todos os valores verdadeiros da variável dependente.

- $ \sum_{i=1}^{n} (y_i - f(x_i))^2 $: Soma dos quadrados dos resíduos, ou seja, o erro do modelo.

- $\sum_{i=1}^{n} (y_i - \bar{y})^2$: Soma total dos quadrados, que representa a variabilidade total nos dados.

\

- Características

1. Valor entre 0 e 1:
  - 1 indica que o modelo explica toda a variância dos dados.
  - 0 indica que o modelo não explica nenhuma variância (equivalente a prever a média sempre).
2. Pode ser negativo se o modelo for pior que prever a média.

\

- Vantagens
1. Fácil interpretação: indica o quão bem o modelo está ajustado aos dados.

\

- Desvantagens
1. Não penaliza erros grandes.
2. Pode ser inflado ao incluir mais variáveis, mesmo que irrelevantes.

- Exemplo Prático
Se \( R^2 = 0.85 \), isso significa que o modelo explica 85% da variância dos dados.

#### Comparando as Métricas

| **Métrica** | **Quando usar** |
|--------------|------------------|
| MAE          | Quando deseja interpretar o erro de forma simples, na mesma escala dos dados. |
| MSE          | Quando erros grandes são mais importantes e precisam ser penalizados. |
| RMSE         | Quando prefere interpretar o erro na mesma escala dos dados, mas com sensibilidade a erros grandes. |
| R² Score     | Quando quer avaliar o ajuste global do modelo e comparar modelos diferentes. |

\

- Conclusão

Compreender as métricas de avaliação é essencial para escolher o melhor modelo para resolver um problema específico. Dependendo do contexto, você pode preferir métricas que penalizem mais os erros grandes (como MSE ou RMSE) ou métricas que forneçam uma interpretação direta e intuitiva (como MAE e R²). Avalie sempre o impacto das decisões baseadas nos resultados das métricas!

#### Previsões com os Dados de Teste

```python
y_pred = reg.predict(X_test)
```

## <font color='yellow'>Classificação</font>

### <font color='yellow'>Métrica para Avaliação de Modelos de Classificação</font>

**Avaliação para Classificação**

Existe uma grande variedade de tarefas de classificação, desde a mais simples, em que temos apenas duas classes, classificação binária, até tarefas mais complexas, como classificação de uma classe, classificação multiclasses, classificação multirrótulos, classificação hierárquica, classificação contínua, classificação por *ranking*, para mencionar as formas mais comuns. Aqui, será abordada a classificação binária, por ser mais simples, mais utilizada e porque alguns algoritmos de Aprendizado de Máquina (AM) induzem modelos apenas para classificação binária. As medidas utilizadas para classificação binária podem ser facilmente expandidas para as outras formas de classificação.

Em uma tarefa de classificação binária, uma das classes, em geral, a de maior interesse, é denominada classe positiva (P) e a outra classe negativa (N). Para facilitar a compreensão dos conceitos, vamos assumir uma tarefa de diagnóstico médico cujas classes positiva e negativa são, respectivamente, doente e saudável. Um modelo preditivo de classificação, aqui chamado de classificador, pode cometer, portanto, dois tipos de erro:

1. Classificação de um exemplo da classe N como da classe P, denominado Falso Positivo (FP), ou alarme falso. Para a tarefa de diagnóstico médico, seria o diagnóstico de um paciente como doente, quando ele está saudável.

2. Classificação de um exemplo da classe P como da classe N, denominado Falso Negativo (FN). Para a tarefa de diagnóstico médico, seria o diagnóstico de um paciente como saudável, quando ele está doente.

Além disso, a classificação correta é chamada Verdadeiro Positivo (VP) para a classe positiva, e Verdadeiro Negativo (VN) para a classe negativa. Uma forma de ilustrar a quantidade de objetos em cada classe, as predições de um modelo preditivo, e os acertos e os erros cometidos, é por meio de uma matriz de confusão, como mostra o exemplo abaixo para um problema de classificação binária, assumindo que a classe 1 é a classe positiva e a classe 2 é a negativa.

\

|                | classe 1 predita | classe 2 predita |
|:--------------:|:---------------:|:---------------:|
| **classe 1 verdadeira** |        25       |        0       |
| **classe 2 verdadeira** |        10       |        40       |

\

De acordo com essa matriz, o conjunto de dados possui 75 objetos, 25 da classe 1 e 50 da classe 2. Todos os 25 objetos da classe 1 foram corretamente classificados na clase 1. Dos objetos da classe 2, 40 foram corretamente classificados na classe 2 e 10 foram incorretamente classificados na classe 1. A matriz de confusão mostra ainda, nos seus elementos, as quantidades de *FP, FN, VP e VN*, que são, respectivamente, 10, 0, 25, 40. Essa relação torna-se mais clara observando as matrizes abaixo, assumindo que a classe 1 é a classe positiva e a classe 2 é a negativa.


\

|                | classe 1 predita | classe 2 predita |
|:--------------:|:---------------:|:---------------:|
| **classe 1 verdadeira** |        25       |        0       |
| **classe 2 verdadeira** |        10       |        40       |

\

|                | classe 1 predita | classe 2 predita |
|:--------------:|:---------------:|:---------------:|
| **classe 1 verdadeira** |        VP       |        FN       |
| **classe 2 verdadeira** |        FP       |        VN       |


\

---

\

**MEDIDAS SIMPLES**

\

Os valores de **FP, FN, VP e VN** são utilizados para definir várias medidas importantes de modo a avaliar o desempenho de classificadores. Duas delas são a Taxa de **FP** e a Taxa de **FN**. A Taxa de **FP**, **TFP**, calculada por meio da Seguinte Equação:

\

$$
\displaystyle
TFP = \frac{FP}{FP + VN}
$$

\

é a proporção de exemplos da classe negativa que são incorretamente classificados como pertencentes à classe positiva. Seu denominador é o número total de exemplos da classe positiva, classificados corretamente ou não. Essa taxa é também chamada erro do tipo I.

A Taxa de FN, **TFN**, calculada por meio da Equação:

\

$$
\displaystyle
TFN = \frac{FN}{FN + VP}
$$

\

é o inverso, ou seja, a proporção de exemplos da classe positiva que são incorretamente classificados como pertencentes à classe negativa. Seu denominador é o número total de exemplos da classe negativa, classificados corretamente ou não. A **TFN** é também chamada de erro do tipo II.

Outra taxa muito utilizada é a taxa de **VP**, **TVP**, mostrada na seguinte Equação:

\

$$
\displaystyle
TVP = recall = revocação = Sensibilidade = \frac{VP}{FN + VP}
$$

\

que indica a quantidade de exemplos da classe positiva correramente classificados nessa classe, ou a porcentagem de objetos positivos que são classificados como positivos. Seu maior valor é obtido quando nenhum objeto positivo deixa de ser classificado na classe positiva, ou seja, todos os exemplos positivos são corretamente classificados. Observe que o denominador da **TVP** é o mesmo da **TFN**. Como quanto maior melhor, ela é chamada de benefício, enquanto a **TFN**, em que quanto maior pior, é chamada, além de erro do tipo II, de custo.

A **TVP** também é chamada de taxa de sensibilidade e de taxa de revocação (recall). Outra medida muito utilizada é o Valor Predito Positivo, **VPP**, mais conhecido pelo termo **precisão (*precision*)**, e calculada pela Equação:

\

$$
\displaystyle
VPP = precisão = precision = \frac{VP}{FP + VP}
$$

\

Seu denominador é formado por todos os objetos classificados na classe positiva, de modo correto ou não.

A precisão retorna a porcentagem de exemplos classificados como positivos que são realmente positivos. Seu maior valor é obtido quando nenhum objeto negativo é incorretamente classificado na classe positiva. Ou seja, todos os objetos que o classificador atribui à classe positiva são dela. De maneira semelhante à **VPP**, mas para a classe negativa, temos o Valor Predito Negativo, **VPN**, mais conhecida como especificidade. Porcentagem de exemplos negativos classificados como negativos. Seu maior valor é obtido quando todos os exemplos negativos são corretamente classificados, ou seja, nenhum exemplo negativo é deixado de fora. A Equação a seguir mostra como esse cálculo é feito:

\

$$
\displaystyle
VPN = especificidade = \frac{VN}{FP + VN}
$$

\

Uma das medidas simples mais utilizadas é a **acurácia**, que retorna quantos objetos foram corretamente classificados. A acurácia pode ser facilmente utilizada em problemas de classificação multiclasses. Seu cálculo é mostrado na Equação abaixo.

\

$$
\displaystyle
Acurácia = \frac{VN + VP}{VN + VP + FN + FP}
$$

\

Ao mesmo tempo que ela apresenta um melhor resumo do desempenho de um classificador, ela omite como os erros (e acertos) estão distribuídos entre as classes. Assim, a classe com mais objetos acaba contribuindo mais no cálculo do valor. Isso faz com que seu uso não seja indicado para dados desbalanceados. Uma de suas variaçoes, a acurácia balanceada, faz com que todas as classes contribuam igualmente no cálculo da acurácia. Essas medidas são simples, podendo ser combinadas para formular medidas mais elaboradas


- **support**

Suporte é o número de ocorrências reais da classe no conjunto de dados especificado. Suporte desbalanceado nos dados de treinamento pode indicar fraquezas estruturais nas pontuações relatadas do classificador e pode indicar a necessidade de amostragem estratificada ou rebalanceamento. O suporte não muda entre os modelos, mas diagnostica o processo de avaliação.

```python
# Separando o conjunto de dados
# X = características (atributos preditivos)
# y = atributo alvo

X = load_iris().data  # Características das flores
y = load_iris().target  # Atributo alvo
```

#### Preparando os Dados

```python
# Dividindo o conjunto de dados em conjunto de treino e teste
# Divisão em dados de treino e teste (80/20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
```

#### Criando o Modelo de Classificação k-NN (Nearest Neighbors Classification)

```python
knn = KNeighborsClassifier(n_neighbors=5)  # Usando 5 vizinhos
```

#### Treinando o Modelo de Classificação k-NN (Nearest Neighbors Classification)

```python
knn.fit(X_train, y_train)
```

#### Previsões com os Dados de Teste

```python
y_pred = knn.predict(X_test)
```

#### Avaliação do Modelo k-NN (Nearest Neighbors Classification)

```python
# Crie um relatório de texto mostrando as principais métricas de classificação.
print(classification_report(y_test, y_pred))
```

```python
print(f'Acurácia: {accuracy_score(y_test, y_pred)}')
print(f'Acurácia Balanceada: {balanced_accuracy_score(y_test, y_pred)}')
print('Matriz de confusao: \n', confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay


# Passando o estimador (o modelo) como parâmetro
ConfusionMatrixDisplay.from_estimator(knn, X_test, y_test)
```

```python
# Passando apenas os valores verdadeiros e preditos
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
from sklearn.metrics import multilabel_confusion_matrix

#Avaliando a matriz de confusão para problemas com mais de duas classes
matrices = multilabel_confusion_matrix(y_test, y_pred)

# para cada classe representada na matriz
for i, matrix in enumerate(matrices):
    tn, fp, fn, tp = matrix.ravel()
    print(f"Classe {i}:")
    print(f'tn: {tn}')
    print(f'fp: {fp}')
    print(f'fn: {fn}')
    print(f'tp: {tp}')
    print("-" * 20)
```

```python
# Calcular probabilidades para a curva ROC
from sklearn.preprocessing import label_binarize

y_test_binarized = label_binarize(y_test, classes=[0, 1, 2])
y_pred_proba = knn.predict_proba(X_test)

# Gerar curva ROC
fpr = {}
tpr = {}
roc_auc = {}
for i in range(3):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plotar a curva ROC
plt.figure()
for i, color in zip(range(3), ['blue', 'red', 'green']):
    plt.plot(fpr[i], tpr[i], label=f'Classe {iris.target_names[i]} (AUC = {roc_auc[i]:.2f})', color=color)

plt.plot([0, 1], [0, 1], 'k--', label='Aleatório')
plt.xlabel('Taxa de Falsos Positivos (TFP)')
plt.ylabel('Taxa de Verdadeiros Positivos (TVP)')
plt.title('Curva ROC')
plt.legend()
plt.show()
```

#### Prevendo a classe com novos dados (Deploy)

```python
# Prevendo a classe do novo objeto
# input = [Comprimento da sépala, Largura da sépala, Comprimento da pétala, Largura da pétala]
input_iris = np.array([[5.1, 3.5, 1.4, 0.2]])
pred_score = knn.predict(input_iris)

print(f'Input = {input_iris}')
print(f'Classe Prevista = {pred_score[0]}')
print(f'Classe Prevista = {iris.target_names[pred_score[0]]}')
```

```python
new_prediction_proba = knn.predict_proba(input_iris)

print(f'Classe prevista: {iris.target_names[pred_score[0]]}')
print(f'Probabilidades por classe: {new_prediction_proba}')
```
